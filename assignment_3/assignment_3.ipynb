Quartic Oscillator Hamiltonian:
$$ H(p, q) = \frac{p^2}{2} + \frac{q^4}{4} = h.$$
The momentum $p$ and position $q$ can be expressed in terms of the conserved energy $h$ and the conjugated variable as
$$ p = \pm \sqrt{2 \left( h - \frac{q^4}{4} \right)} $$
and
$$ q = \pm \sqrt{2} \left( h - \frac{p^2}{2} \right)^{1/4}.$$

```python
import numpy as np
import scipy as sp
import scipy.integrate as sp_int
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from functools import partial
```

```python
h = lambda q, p: p**2 / 2 + q**4 / 4
p = lambda q, h: np.sqrt(2 * (h - q**4 / 4))
q = lambda p, h: np.sqrt(2) * (h - p**2 / 2)**(1/4)
```

# Exercise 1

```python
plt.rcParams.update({'font.size': 18})
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1, 1, 1)

q_space = np.linspace(-5, 5, 1000)
p_space = np.linspace(-15, 15, 1000)

Q_space, P_space = np.meshgrid(q_space, p_space)
H = h(Q_space, P_space)

energy_levels = np.linspace(0, 100, 21)
cs = ax.contour(q_space, p_space, H, levels=energy_levels, colors='black', linestyles='--')
csf = ax.contourf(q_space, p_space, H, levels=energy_levels, cmap=plt.cm.RdYlBu_r)

cbar = fig.colorbar(csf)
cbar.ax.set_ylabel('$h$')
cbar.add_lines(cs)

ax.set_xlabel('$q$')
ax.set_ylabel('$p$')

fig.savefig('ex1_phase_plot.png', bbox_inches='tight')
```
